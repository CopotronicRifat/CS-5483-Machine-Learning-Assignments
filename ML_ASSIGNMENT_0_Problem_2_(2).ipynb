```python
# Take input image from local drive
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```python
# Load image

import cv2   
img = cv2.imread(fn)
```

```python
# Convert BGR image to RGB image

b,g,r = cv2.split(img) 
rgb = cv2.merge([r,g,b])
copy4 = rgb
```

```python
# Print image array values just to visualize

print(rgb)
```

```python
# Problem 2(2)

for i in range(-25+int(img.shape[0]/2), 25+int(img.shape[1]/2)):
    for j in range(-25+int(img.shape[0]/2), 25+int(img.shape[1]/2)):
        for k in range(3):
            copy4[i][j][k] = 0
```

```python
# Show and Save Image

from matplotlib import pyplot as plt
plt.imshow(copy4)
plt.xticks([]), plt.yticks([])   # to hide tick values on X and Y axis
plt.show()
plt.savefig("output2.png")
```
