**Answer 1.2:**

**a.** From the visualization from Train Data and Test Data, it is obvious that the relationship between feature and level is not linear.

**b.** Yes. We do need to feature engineering to add non-linearity. 

***i.*** We can gradually increase the order of polynomial to get the non-linearity to see what fits best for the train data. 


***ii.*** We will start from the following equation as our hypothesis function:

$h(x) = b_0 + b_1*x + b_2*x^2$

and if this equation doesn't fit these data then we can increase the order of the polynomial and use the following equations:

$h(x) = b_0 + b_1*x + b_2*x^2 + b_3*x^3$

$h(x) = b_0 + b_1*x + b_2*x^2 + b_3*x^3 + b_4*x^4$

and so on. We will fix which order of polynomial fits best for the training data.







Plot of each of the polynomial equations:

```python
# random initialization of the parameters

theta = []
import random
for i in range(5):
    num = random.randint(0, 10)
    theta.append(num)

print(theta)
```

```python
# plot of 2nd Order Polynomial
m = len(train_y)
h = []
for i in range(m):
    temp = theta[0] + theta[1]*train_x[i] + theta[2]*train_x[i]**2
    h.append(temp)

plt.scatter(train_x, train_y, marker = "x", label = "Training Set")
plt.scatter(train_x, h, marker = "o", label = "Hypothesis of 2nd Order")
plt.title('2nd Order Polynomial')
plt.xlabel('x_train')
plt.ylabel('hypothesis of 2nd order')
plt.legend()
plt.show
```

```python
# plot of 3rd Order Polynomial
m = len(train_x)
h = []
for i in range(m):
    temp = theta[0] + theta[1]*train_x[i] + theta[2]*train_x[i]**2 + theta[3]*train_x[i]**3
    h.append(temp)

plt.scatter(train_x, train_y, marker = "x", label = "Training Set")
plt.scatter(train_x, h, marker = "o", label = "Hypothesis of 3rd Order")

plt.title('3rd Order Polynomial')
plt.xlabel('x_train')
plt.ylabel('hypothesis of 3rd order')
plt.legend()
plt.show
```

```python
# plot of 4th Order Polynomial

h = []
for i in range(m):
    temp = theta[0] + theta[1]*train_x[i] + theta[2]*train_x[i]**2 + theta[3]*train_x[i]**3 + theta[4]*train_x[i]**4
    h.append(temp)

plt.scatter(train_x, train_y, marker = "x", label = "Training Set")
plt.scatter(train_x, h, marker = "o", label = "Hypothesis of 4th Order")
plt.title('4th Order Polynomial')
plt.xlabel('x_train')
plt.ylabel('hypothesis of 4th order')
plt.legend()
plt.show
```

```python
# parameter initialization
m = len(train_x)
alpha = 0.01
```

```python
theta = []
import random
for i in range(4):
    num = random.randint(0, 10)
    theta.append(num)

print(theta)
```

**Polynomial Linear Regression Implementation** 

```python
# parameter initialization
J = 1000
n = 0 

# I run this linear regression model 1000 times
for i in range(1000):
  print("Iteration number: ",n+1)
  n = n+1
  # hypothesis function
  h = []
  for i in range(m):
    temp = theta[0] + theta[1]*train_x[i] + theta[2]*train_x[i]**2 + theta[3]*train_x[i]**3
    h.append(temp)
  print("Hypothesis function value is: ",h)

  # cost function
  error_sum = 0 
  for i in range(m):
    error_sum = error_sum + (h[i] - train_y[i])**2

  J = (1/(2*m))*error_sum

  print("Cost function is:",J)

  # gradient descent
  temp0 = 0
  for i in range(m):
    temp0 = temp0 + (h[i] - train_y[i])

  theta[0] = theta[0] - (alpha/m)*temp0

  temp1 = 0
  for i in range(m):
    temp1 = temp1 + (h[i] - train_y[i])*train_x[i]

  theta[1] = theta[1] - (alpha/m)*temp1

  temp2 = 0
  for i in range(m):
    temp2 = temp2 + (h[i] - train_y[i])*2*train_x[i]

  theta[2] = theta[2] - (alpha/m)*temp2

  temp3 = 0
  for i in range(m):
    temp3 = temp3 + (h[i] - train_y[i])*3*train_x[i]

  theta[3] = theta[3] - (alpha/m)*temp3


  print("New parameter value is: ",theta)
  

```

```python
print("predicted hypothesis is:",h)
plt.scatter(train_x, train_y, marker = "x", label = "Train Dataset")
plt.scatter(train_x, h, marker = "o", label = "Hypothesis Function after Gradient Descend Applied")

plt.title('Comparison of Training Data vs Predicted Value')
plt.xlabel('Data')
plt.ylabel('Level')

plt.legend()
plt.show
```
