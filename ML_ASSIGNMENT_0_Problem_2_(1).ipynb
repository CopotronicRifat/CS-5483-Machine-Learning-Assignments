```python
# Take input image from local drive
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```python
# Load image

import cv2   
img = cv2.imread(fn)
```

```python
# Convert BGR image to RGB image

b,g,r = cv2.split(img) 
rgb = cv2.merge([r,g,b])
```

```python
# Print image array values just to visualize

print(rgb)
```

```python
# Problem 2(1)

import numpy as np

point1 = np.array((255, 0, 0))
point2 = np.array((0, 255, 0))
point3 = np.array((0, 0, 255))

for i in range(rgb.shape[0]):
    for j in range(rgb.shape[1]):
        dist = []
        dist.append(np.linalg.norm(rgb[i][j] - point1))
        dist.append(np.linalg.norm(rgb[i][j] - point2))
        dist.append(np.linalg.norm(rgb[i][j] - point3))

        if dist[0]<=dist[1] and dist[1]<=dist[2]:
            rgb[i][j] = point1
        elif dist[1]<=dist[0] and dist[1]<=dist[2]:
            rgb[i][j] = point2
        else:
            rgb[i][j] = point3
```

```python
# Show and Save Image

from matplotlib import pyplot as plt
plt.xticks([]), plt.yticks([])
plt.imshow(rgb)
plt.show()
plt.savefig("output1.png")
```
